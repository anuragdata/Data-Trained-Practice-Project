```python
for column in hr.columns:
    if hr[column].dtype==object:
        print((column)  + ' : ' +  (hr[column].unique()))
        print(hr[column].value_counts())
        print('\n')                                     #viewing object type data and its value count
```

```python
hr['Gender'] = hr['Gender'].replace({'Female': 0, 'Male': 1})
hr['OverTime'] = hr['OverTime'].replace({'No': 0, 'Yes': 1})    
hr['BusinessTravel'] = hr['BusinessTravel'].replace({'Non-Travel': 0, 'Travel_Rarely': 1, 'Travel_Frequently': 2})
hr['Attrition']=hr['Attrition'].replace({'No': 0, 'Yes': 1})        #Converting object type data column into int.
```

```python
 # Drop single-value columns and id columns
df = hr.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1) #droping columns which have single value.
```

```python
df.head()
```

```python
def dummy(df, column):
    df=df.copy()
    dummy=pd.get_dummies(df[column])
    df=pd.concat([df,dummy],axis=1)
    df=df.drop(column,axis=1)
    return df
```

```python
for column in ['Department', 'EducationField', 'JobRole', 'MaritalStatus']:
        df = dummy(df, column=column)
```

```python
df
```

```python
y=df['Attrition']
x=df.drop(['Attrition'],axis=1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
import warnings
import joblib
warnings.filterwarnings('ignore')
```

```python
lr=LogisticRegression()
```

```python
maxr2=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    acc=accuracy_score(y_test,predlr)
    if acc>maxr2:
        maxr2=acc
        maxRS=i
print("Best accuracy is ",maxr2," on Random_state ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=130)
```

```python
model=[KNeighborsClassifier(),DecisionTreeClassifier(),GaussianNB(),SVC()]
for m in model:
    m.fit(x_train,y_train)
    predm=m.predict(x_test)
    print("accuracy score of ", m , "is:")
    print(accuracy_score(y_test,predm))
```

```python
from sklearn.model_selection import cross_val_score
model=[KNeighborsClassifier(),DecisionTreeClassifier(),GaussianNB(),SVC(),LogisticRegression()]
for i in model:
    score=cross_val_score(i,x,y,cv=5)
    print("score of ", i , "is:")
    print(score)
    print(score.mean())
```

Here Score of LogisticRegression is better than other, so our best model is Logistic Regression

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
```

```python
parameters={'alpha':[0.0001,0.001,0.01,.1,1,10],'random_state':list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

GridSearchCV

```python
ls=Lasso(alpha=0.001,random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
pred_ls=ls.predict(x_test)
lss=r2_score(y_test,pred_ls)
lss
```

```python
parameters={'criterion':['gini', 'entropy'],'max_features':['auto', 'sqrt', 'log2']}
rf=RandomForestClassifier()
clf=GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)
print(clf.best_score_)
print(clf.best_params_)
```

Ensemble- RandomForest

```python
rf=RandomForestClassifier(criterion='gini',max_features='auto')
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
rfscore=cross_val_score(rf,x,y,cv=5)
rfscore=rfscore.mean()
print('cross val:',rfscore*100)
```

```python
import joblib
joblib.dump(rf,'hratt.obj') #dumping the best model
```
