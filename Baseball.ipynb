```python
x=df.drop(columns=['W'])
y=df['W']
```

```python
for i in x.columns:
    if x[i].skew()>0.5:
        x[i]=np.cbrt(x[i])
    if x[i].skew()<=0.5:
        x[i]=np.cbrt(x[i])
x.skew()                           #Removing Skewness
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso,Ridge
```

```python
lr=LinearRegression()
```

```python
maxr2=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    rscore=r2_score(y_test,predlr)
    if rscore>maxr2:
        maxr2=rscore
        maxRS=i
print("Best r score is ",maxr2," on Random_state ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=151)
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
rscore=r2_score(y_test,predlr)
print(rscore)
```

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(lr,x,y,cv=5)
cv_mean=cv_score.mean()
print(cv_mean)
    
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=predlr,color='r')
plt.plot(x_test,x_test,color='b')
plt.xlabel('actual value')
plt.ylabel('predicted value')
plt.title('fitting of value')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters={'alpha':[0.0001,0.001,0.01,.1,1,10],'random_state':list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)
print(clf.best_params_)
```

```python
ls=Lasso(alpha=1,random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
pred_ls=ls.predict(x_test)
lss=r2_score(y_test,pred_ls)
lss
```

```python
cv_score=cross_val_score(ls,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
parameters={'criterion':['mse', 'mae'],'max_features':['auto', 'sqrt', 'log2']}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)
print(clf.best_score_)
print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion='mae',max_features='auto')
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_dec=rf.predict(x_test)
rfs=r2_score(y_test,pred_dec)
print('rfs score:', rfs*100)
rfscore=cross_val_score(rf,x,y,cv=5)
rfscore=rfscore.mean()
print('cross val:',rfscore*100)
```

Lasso Regression has best score, so here this regression model is picked for prediction.

```python
import joblib
joblib.dump(ls,'baseball.obj') #dumping the best model
```
