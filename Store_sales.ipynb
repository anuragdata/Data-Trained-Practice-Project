```python
dummies=pd.get_dummies(train_new.Location_Type)
```

```python
df=pd.concat([df,dummies],axis='columns')
```

```python
df.drop(['Location_Type','L5'],axis='columns',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['dayOfWeek'] = le.fit_transform(df['dayOfWeek'])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Discount'] = le.fit_transform(df['Discount'])
```

```python
df.dtypes
```

```python
df['year'].replace({2018:0,2019:1},inplace=True)
df
```

```python
y=df['Sales']
x=df.drop(['ID','Sales','#Order'],axis='columns')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso,Ridge
```

```python
lr=LinearRegression()
```

```python
maxr2=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=i)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    rscore=r2_score(y_test,predlr)
    if rscore>maxr2:
        maxr2=rscore
        maxRS=i
print("Best r score is ",maxr2," on Random_state ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=106)
```

```python
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
rscore=r2_score(y_test,predlr)
print(rscore)
```
